##  Q1. Which country has the highest total fertility rate in 2019?

```python
# create fertility data frame
hfr = hnp_df[['Country Name', 'Country Code', 'Indicator Name', '2019']]
hfr = hfr[hfr['Indicator Name'] == 'Fertility rate, total (births per woman)']
hfr.head()
```

```python
# determine the highest total fertility rate in 2019 (using sort_values)
hfr.sort_values(by=['2019'], ascending = False).head(1)
```

```python
# determine the highest total fertility rate in 2019 (using max)
hfr[hfr['2019'] == hfr['2019'].max()]
```

###  **Niger** has the highest total fertility rate of `6.824` total births per woman.
